```python
y=df['sentiment']
x=df['stopword_removal']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
nb=MultinomialNB()
log=LogisticRegression()
dtc=DecisionTreeClassifier()
```

```python
x_dtm=vect.fit_transform(x)
```

```python
x_dtm.shape
```

```python
def tweet_sentence(score): 
    if score > 0.2: 
        return 1
    elif score < -0.2:
        return -1
    else:
        return 0       
df['class']=df['sentiment'].apply(lambda x: tweet_sentence(x))
```

```python
y=df['class']
```

```python
y
```

```python
algorithms=[nb,log,dtc]
names=['MultinomialNB','LogisticRegression','DecisionTreeClassifier']

def classification():
    accuracy=[]
    precision=[]
    recall=[]
    f1=[]
    for i in range(len(algorithms)):
        algorithms[i].fit(x_dtm,y)
    for i in range(len(algorithms)):
        accuracy.append(accuracy_score(y,algorithms[i].predict(x_dtm)))
        precision.append(precision_score(y,algorithms[i].predict(x_dtm)))
        recall.append(recall_score(y,algorithms[i].predict(x_dtm)))
        f1.append(f1_score(y,algorithms[i].predict(x_dtm)))
        
    met=pd.DataFrame(columns=['Accuracy','Precision','Recall','F1'], index=names)
    met['Accuracy']=accuracy
    met['Precision']=precision
    met['Recall']=recall
    met['F1']=f1
    return met    
```
